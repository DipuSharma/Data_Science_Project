## Fashion Classification

```python
from keras.datasets import fashion_mnist
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
```

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
X_train[0]
```

```python
class_label = ["T-Shirt/Top", "Trouser", "Pullover", "Dress", "Coat", "Sandel", "Shirt", "Sneaker", "Bag", "Ankle Boot"]
```

```python
plt.imshow(X_train[0], cmap="Greys");
```

```python
plt.figure(figsize=(5,5))
plt.imshow(X_train[0], cmap="Greys")
plt.show()
```

```python
plt.figure(figsize=(20,20))
for i in range(25):
  plt.subplot(5, 5, i+1)
  plt.imshow(X_train[i], cmap="Greys")
  plt.axis('off')
plt.show()
```

## Feature Scalling

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
len(X_train[0])
```

```python
y_test[0]
```

## Building Neural Network

```python
from tensorflow import keras
model = keras.models.Sequential([ keras.layers.Flatten(input_shape=[28, 28]), keras.layers.Dense(units=32, activation='relu'), keras.layers.Dense(units=10, activation='softmax')                               
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=1)
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test, y_test)
```

```python
pred = model.predict(X_test)
```

```python
pred[0]
```

```python
pred[0].round(2)
```

```python
np.argmax(pred[0].round(2))
```

```python
y_test[0]
```

```python
plt.figure(figsize=(20, 20))
for i in range(25):
  plt.subplot(5, 5, i+1)
  plt.imshow(X_test[i], cmap='Greys')
  plt.axis('off')
  plt.title("Actuals = {} \n Predict = {}".format(class_label[y_test[i]], class_label[np.argmax(pred[i])]))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, [np.argmax(i) for i in pred])
```

```python
plt.figure(figsize=(10, 8))
snb.heatmap(cm, annot=True, fmt='d')
```

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test, [np.argmax(i) for i in pred], target_names=class_label)
```

```python
print(cr)
```
